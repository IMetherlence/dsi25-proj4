# Data Cleaning for Training and Test Datasets

## Import Libraries

```python
import pandas as pd
from datetime import date
from datetime import datetime
```

## Load Train  Data into Memory

```python
data_train = pd.read_csv("../assets/train.csv")
data_test = pd.read_csv("../assets/test.csv")
```

# Training Dataset:

## Preliminary Data Exploration

Lets have a brief look at our training dataset first

```python
data_train.head()
```

```python
data_train.describe()
```

```python
data_train.info()
```

We can see 12 columns with our Target Variable being the WnvPresent column. Let's continue data cleaning.

## Check for Null or Missing Values

```python
data_train.isnull().sum()
```

Appears we have no null values to deal with so our data should be clean.

## Standardize our Column Names to lower case separated with underscores

Let's standardize our column formatting to lower case letters with underscores separating spaces.

```python
data_train.columns = data_train.columns.str.lower()
data_train.info()
# There are no spaces in the column names thus no need for an additional replace method call
```

## Creating our year, month and date columns

We do note that the data column is in the format YYYY-MM-DD. We need to split them up for use together with our seasonality data from our weather.csv. I will split this column into years, month and date first.

```python
# Converting our date column into a datetime object before retrieving the year from the isocalendar method
data_train['year'] = [datetime.date(datetime.strptime(x, '%Y-%m-%d')).isocalendar()[0] for x in data_train['date']]
data_train.head()
```

```python
# Converting our date column into a datetime object before using the month method to retrieve the month
data_train['month'] = [datetime.date(datetime.strptime(x, '%Y-%m-%d')).month for x in data_train['date']]
data_train.tail()
```

```python
# Converting our date column into a datetime object before using the day method to retrieve the day of the month
data_train['day_of_month'] = [datetime.date(datetime.strptime(x, '%Y-%m-%d')).day for x in data_train['date']]
data_train.head()
```

```python
# Converting our date column into a datetime object before retrieving the day of week from the isocalendar method
data_train['day_of_week'] = [datetime.date(datetime.strptime(x, '%Y-%m-%d')).isocalendar()[2] for x in data_train['date']]
data_train.head()
```

```python
# Converting our date column into a datetime object before retrieving the week of year from the isocalendar method
data_train['week_of_year'] = [datetime.date(datetime.strptime(x, '%Y-%m-%d')).isocalendar()[1] for x in data_train['date']]
data_train.head()
```
